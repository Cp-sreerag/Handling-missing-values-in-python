## Filling Null values


```python
df2 = df.fillna(value=0)
df2
```

```python
df2.isnull().sum().sum()
```

```python
df3 = df.fillna(value=5)
df3
```

### Filling Null values with previous values

```python
df4 = df.fillna(method='pad')
df4
```

```python
df4.isnull().sum().sum()
```

### Filling Null values with coming values

```python
df5 = df.fillna(method='bfill')
df5
```

### Filling Null values with previous values by columnwise

```python
df6 = df.fillna(method='pad',axis=1)
df6
```

```python
df7 = df.fillna(method='bfill',axis=1)
df7
```

### Filling different values in null in different columns

```python
df8 = df.fillna({'society':'abcd',
                  'balcony':'defg'})
df8
```

### Filling null value with the mean of a column

```python
df9 = df.fillna(value=df['balcony'].mean())
df9
```

```python
df10 = df.fillna(value=df['balcony'].min())
df10
```

```python
df11 = df.fillna(value=df['balcony'].max())
df11
```

## Droping Null values

```python
df12 = df.dropna()
df12
```

```python
df13 = df.dropna(how='any')
df13
```

```python
df14 = df.dropna(how='all')
df14
```

## Replacing Null values

```python
import numpy as np
```

```python
df15 = df.replace(to_replace=np.nan,value=1234)
df15
```

## Interpolating Null values

```python
df['balcony']=df['balcony'].interpolate(method = 'linear',limit_direction = 'forward')
df['balcony']
```
